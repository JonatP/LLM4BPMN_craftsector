Disclaimer: The numbering of the processes in this file is inconsistent with the thesis. In this file, the processes indices start at 0 (instead of 1) and the last two processes are switched.

```python
org_file = ["1-ExamRegistration", "2-FrontendMergeRequest","3-RestaurantOrderWithQuickDelivery","4-StartupEvaluationVC"]
output_folder = ["1-examregistration","2-frontendmergerequest","3-restaurantorderwithquickdelivery","4-startupevaluationvc"]

def evaluate(process_number, prompting_method):
    bpmn_true = "./data/controlled-environment/3-bpmn/"+org_file[process_number]+".bpmn"
    bpmn_1 = "./output/prompt-engineering/"+prompting_method+"/"+output_folder[process_number]+"/model.bpmn"
    bpmn_2 = "./output/prompt-engineering/"+prompting_method+"/"+output_folder[process_number]+"/model_1.bpmn"
    bpmn_3 = "./output/prompt-engineering/"+prompting_method+"/"+output_folder[process_number]+"/model_2.bpmn"

    ged_1 = compare_bpmn_files(bpmn_true, bpmn_1)
    ged_2 = compare_bpmn_files(bpmn_true, bpmn_2)
    ged_3 = compare_bpmn_files(bpmn_true, bpmn_3)

    return [ged_1, ged_2, ged_3]
```

## Instruction/Zero-shot

```python
prompting_method = "0-instruction"
for i in range(4):
    ged = evaluate(i, prompting_method)
    print(prompting_method, " process ", i, ": ", ged[0], ged[1], ged[2])
```

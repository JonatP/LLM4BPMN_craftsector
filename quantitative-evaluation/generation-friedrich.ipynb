```python
import sys
import os
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)
from pipeline_final import speech_to_model
```

```python
def generate_bpmn(folder):
    input_path = "./data/friedrich/"+folder+"/1-speech/"
    for file in os.listdir(input_path):
        input_file = input_path+file
        base_name = os.path.splitext(file)[0]
        output_folder = "./output/quantitative-evaluation/"+folder+"/"+base_name
        speech_to_model(input_file, output_folder)
```
